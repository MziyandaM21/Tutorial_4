### Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
N = 100
t = np.linspace(0,4*np.pi,N)
y = np.sin(t)
data = 3.0*y  + np.random.randn(t.size)
datafft = np.real(np.fft.fft(data))
dataifft = np.real(np.fft.ifft(datafft))
npoly = 10
A = np.zeros([t.size,npoly])
A[:,0] = 1.0
for i in range(1,npoly):
    A[:,i]=A[:,i-1]*t

A = np.matrix(A)
d = np.matrix(data).transpose()
lhs = A.transpose()*A
rhs = A.transpose()*d
fitp = np.linalg.inv(lhs)*rhs
pred = A*fitp
plt.plot(t,data,'.', label='Data')
plt.plot(t,pred,'r', label='The Fit')
plt.plot(t,dataifft,'*',label = 'FFT of data')
plt.legend()
plt.show()
```

### Problem 2

```python
def simulate_lorentzian(t,a=1.0,b=0.25,amp=1,cent=0):
    dat=(a/(b+(t-cent)**2))*amp
    dat+=np.random.randn(t.size)
    return dat
    
#get a trial step. Take gaussian random numbers and scale by an input vector
def get_trial_offset(a):
    return a*np.random.randn(a.size)

class Lorentzian:
    def __init__(self,t,a=0,b=0,amp=0,cent=0,offset=0):
        self.t=t
        self.err=np.ones(t.size)
        self.a=a
        self.b=b
        self.amp=amp
        self.cent=cent
        self.offset=offset
        self.y=simulate_lorentzian(t,a,b,amp,cent)+offset
    def get_chisq(self,vec):
        a=vec[0]
        b=vec[0]
        amp=vec[1]
        cent=vec[2]
        off=vec[3]

        pred=off+amp*(self.a/(self.b + (self.t-cent)**2))
        chisq=np.sum((self.y-pred)**2/self.err**2)
        return chisq

def run_mcmc(data,start_pos,nstep,scale=None):
    nparam=start_pos.size
    params=np.zeros([nstep,nparam+1])
    params[0,0:-1]=start_pos
    cur_chisq=data.get_chisq(start_pos)
    cur_pos=start_pos.copy()
    if scale.all()==None:
        scale=np.ones(nparam)
    for i in range(1,nstep):
        new_pos=cur_pos+get_trial_offset(scale)
        new_chisq=data.get_chisq(new_pos)
        if new_chisq<cur_chisq:
            accept=True
        else:
            delt=new_chisq-cur_chisq
            prob=np.exp(-0.5*delt)
            if np.random.rand()<prob:
                accept=True
            else:
                accept=False
        if accept: 
            cur_pos=new_pos
            cur_chisq=new_chisq
        params[i,0:-1]=cur_pos
        params[i,-1]=cur_chisq
    return params
```

```python
t=np.arange(-5,5,0.01)
dat=Lorentzian(t,amp=2.5)
guess=np.array([0.3,1.2,0.3,-0.2])
scale=np.array([0.1,0.1,0.1,0.1])
nstep=100000
chain=run_mcmc(dat,guess,nstep,scale)
nn=int(np.round(0.2*nstep))
chain=chain[nn:,:]
param_true=np.array([dat.a,dat.b,dat.amp,dat.cent,dat.offset])
for i in range(0,param_true.size):
    val=np.mean(chain[:,i])
    scat=np.std(chain[:,i])
    print ([param_true[i],val,scat])
```

```python
n=300
rho=np.zeros(n)
rho[n//3:(2*n//3)]=1
v=1.0
dx=1.0
x=np.arange(n)*dx
plt.ion()
plt.clf()
plt.axis([0,n,0,1.1])
plt.plot(x,rho)
plt.draw()
plt.savefig('advect_initial_conditions.png')
dt=1.0
big_rho=np.zeros(n+1)
big_rho[1:]=rho
del rho  
for step in range(0,250):
    big_rho[0]=big_rho[-1]
    drho=big_rho[1:]-big_rho[0:-1]
    big_rho[1:]=big_rho[1:]-v*dt/dx*drho
    plt.clf()
    plt.axis([0,n,0,1.1])
    plt.plot(x,big_rho[1:])
    plt.draw()
```
